```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("features.csv",header=None)
df.head()
```

```python
data = np.array(df)
data.shape



```

```python
val = []
for i in range(data.shape[0]):
    lst = []
    for j in range(6):
       if(j<4): 
        lst.append(data[i][j])
       else:
        if(data[i][j]<=4):   
         lst.append(0)
        else:
         lst.append(1)
    val.append(lst)     
```

```python
arr = np.array(val)
```

```python
arr.shape
```

```python
df = pd.DataFrame(arr)
df.columns = ["Activity","Mobility","Complexity","NSI","Valence","Arousal"]
df.head(40)
```

```python
df.to_csv('features_new.csv', header=False, index=False)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
n_df = df
n_df.head()
```

```python
n_df = pd.DataFrame(scaler.fit_transform(n_df))
```

```python
n_df.head()
n_df.columns = ["Activity","Mobility","Complexity","NSI","Valence","Arousal"]
```
